```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = 'outputs/all_values_sf.csv'
df = pd.read_csv(filepath)

fig, [ax1, ax2, ax3, ax4] = plt.subplots(4, 1, layout='constrained', sharex=True, figsize=(12,24))
ax1.errorbar('power', 'betap', yerr='betap_error', data=df, color='black')
ax1.set_ylabel('$\\beta_p$', fontsize=14)
ax1.grid()
ax2.errorbar('power', 'ne0', yerr='ne0_error', data=df, color='red')
# ax2.set_ylim([0, 0.012])
ax2.set_ylabel('$n_{e0}$', fontsize=14)
ax2.grid()
ax3.errorbar('power', 'taue', yerr='taue_error', data=df)
# ax3.set_ylim([0, 2])
ax3.set_ylabel('$\\tau_e$', fontsize=14)
ax3.grid()
ax4.errorbar('power', 'te0', yerr='te0_error', data=df, color='orange')
# ax4.set_ylim([0, 0.24])
ax4.set_ylabel('$t_{e0}$', fontsize=14)
ax4.grid()
ax4.set_xlabel('Power (MW)', fontsize=14)
fig.suptitle('Power vs. $\\beta_p$, $n_{e0}$, $\\tau_e$ and $t_{e0}$', fontsize=18)
plt.show()
```
